```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the dataset
# Replace with your dataset path or create a small sample dataset
data = {
    "Month": ["Jan-2023", "Jan-2023", "Feb-2023", "Feb-2023", "Mar-2023", "Mar-2023"],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Sales": [120, 90, 150, 130, 200, 170],
    "Revenue": [3000, 2700, 3750, 3900, 5000, 5100],
}
df = pd.DataFrame(data)

# Step 2: Explore the dataset
print("Dataset Overview:")
print(df.head())

# Total revenue and sales
total_sales = df["Sales"].sum()
total_revenue = df["Revenue"].sum()
print(f"\nTotal Sales: {total_sales}, Total Revenue: ${total_revenue}")

# Step 3: Analyze data with NumPy
mean_sales = np.mean(df["Sales"])
median_sales = np.median(df["Sales"])
std_dev_sales = np.std(df["Sales"])
print(f"\nSales Statistics:\nMean: {mean_sales}, Median: {median_sales}, Std Dev: {std_dev_sales}")

# Step 4: Visualize the data
# Line plot: Monthly sales trends for each product
plt.figure(figsize=(10, 6))
for product in df["Product"].unique():
    product_data = df[df["Product"] == product]
    plt.plot(product_data["Month"], product_data["Sales"], label=f"Product {product}", marker='o')

plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

# Bar plot: Total revenue by product
revenue_by_product = df.groupby("Product")["Revenue"].sum()
revenue_by_product.plot(kind="bar", color=["blue", "orange"], figsize=(8, 6))
plt.title("Total Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()

# Pie chart: Sales contribution by product
sales_by_product = df.groupby("Product")["Sales"].sum()
sales_by_product.plot(kind="pie", autopct="%1.1f%%", startangle=90, figsize=(6, 6))
plt.title("Sales Contribution by Product")
plt.ylabel("")  # Hides the ylabel for better visualization
plt.show()
```
